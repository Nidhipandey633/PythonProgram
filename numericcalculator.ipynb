```python
class NumericCalculator:
    def add(self, a, b):
        return a + b

    def multiply(self, a, b):
        return a * b

    def subtract(self, a, b):
        return a - b

    def divide(self, a, b):
        if b == 0:
            return "Error: Division by zero"
        return a / b

    def modulus(self, a, b):
        return a % b

    def simple_interest(self, p, r, t):
        return (p * r * t) / 100
```

```python
# Create an instance of the NumericCalculator
calculator = NumericCalculator()

# Demonstrate using the functions
print(f"Addition: 5 + 3 = {calculator.add(5, 3)}")
print(f"Multiplication: 5 * 3 = {calculator.multiply(5, 3)}")
print(f"Subtraction: 5 - 3 = {calculator.subtract(5, 3)}")
print(f"Division: 10 / 2 = {calculator.divide(10, 2)}")
print(f"Division by zero: {calculator.divide(10, 0)}")
print(f"Modulus: 10 % 3 = {calculator.modulus(10, 3)}")
print(f"Simple Interest (P=1000, R=5, T=2): {calculator.simple_interest(1000, 5, 2)}")
```
