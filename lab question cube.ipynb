```python
import math

def cube_curved_surface_area(side):
    """Calculates the curved surface area of a cube (4 sides)."""
    return 4 * side**2

def cube_total_surface_area(side):
    """Calculates the total surface area of a cube (6 sides)."""
    return 6 * side**2

def cube_volume(side):
    """Calculates the volume of a cube."""
    return side**3

def cuboid_curved_surface_area(length, width, height):
    """Calculates the curved surface area of a cuboid (lateral surface area)."""
    return 2 * height * (length + width)

def cuboid_total_surface_area(length, width, height):
    """Calculates the total surface area of a cuboid."""
    return 2 * (length * width + length * height + width * height)

def cuboid_volume(length, width, height):
    """Calculates the volume of a cuboid."""
    return length * width * height

def cylinder_curved_surface_area(radius, height):
    """Calculates the curved surface area of a cylinder."""
    return 2 * math.pi * radius * height

def cylinder_total_surface_area(radius, height):
    """Calculates the total surface area of a cylinder."""
    return 2 * math.pi * radius * (radius + height)

def cylinder_volume(radius, height):
    """Calculates the volume of a cylinder."""
    return math.pi * radius**2 * height

def cone_curved_surface_area(radius, slant_height):
    """Calculates the curved surface area of a cone."""
    return math.pi * radius * slant_height

def cone_total_surface_area(radius, slant_height):
    """Calculates the total surface area of a cone."""
    return math.pi * radius * (radius + slant_height)

def cone_volume(radius, height):
    """Calculates the volume of a cone."""
    return (1/3) * math.pi * radius**2 * height

def sphere_curved_surface_area(radius):
    """Calculates the curved surface area of a sphere (same as total surface area)."""
    return 4 * math.pi * radius**2

def sphere_total_surface_area(radius):
    """Calculates the total surface area of a sphere."""
    return 4 * math.pi * radius**2

def sphere_volume(radius):
    """Calculates the volume of a sphere."""
    return (4/3) * math.pi * radius**3

def hemisphere_curved_surface_area(radius):
    """Calculates the curved surface area of a hemisphere."""
    return 2 * math.pi * radius**2

def hemisphere_total_surface_area(radius):
    """Calculates the total surface area of a hemisphere."""
    return 3 * math.pi * radius**2

def hemisphere_volume(radius):
    """Calculates the volume of a hemisphere."""
    return (2/3) * math.pi * radius**3

# Write the module content to a file
with open('geometry_calculations.py', 'w') as f:
    f.write('''
import math

def cube_curved_surface_area(side):
    """Calculates the curved surface area of a cube (4 sides)."""
    return 4 * side**2

def cube_total_surface_area(side):
    """Calculates the total surface area of a cube (6 sides)."""
    return 6 * side**2

def cube_volume(side):
    """Calculates the volume of a cube."""
    return side**3

def cuboid_curved_surface_area(length, width, height):
    """Calculates the curved surface area of a cuboid (lateral surface area)."""
    return 2 * height * (length + width)

def cuboid_total_surface_area(length, width, height):
    """Calculates the total surface area of a cuboid."""
    return 2 * (length * width + length * height + width * height)

def cuboid_volume(length, width, height):
    """Calculates the volume of a cuboid."""
    return length * width * height

def cylinder_curved_surface_area(radius, height):
    """Calculates the curved surface area of a cylinder."""
    return 2 * math.pi * radius * height

def cylinder_total_surface_area(radius, height):
    """Calculates the total surface area of a cylinder."""
    return 2 * math.pi * radius * (radius + height)

def cylinder_volume(radius, height):
    """Calculates the volume of a cylinder."""
    return math.pi * radius**2 * height

def cone_curved_surface_area(radius, slant_height):
    """Calculates the curved surface area of a cone."""
    return math.pi * radius * slant_height

def cone_total_surface_area(radius, slant_height):
    """Calculates the total surface area of a cone."""
    return math.pi * radius * (radius + slant_height)

def cone_volume(radius, height):
    """Calculates the volume of a cone."""
    return (1/3) * math.pi * radius**2 * height

def sphere_curved_surface_area(radius):
    """Calculates the curved surface area of a sphere (same as total surface area)."""
    return 4 * math.pi * radius**2

def sphere_total_surface_area(radius):
    """Calculates the total surface area of a sphere."""
    return 4 * math.pi * radius**2

def sphere_volume(radius):
    """Calculates the volume of a sphere."""
    return (4/3) * math.pi * radius**3

def hemisphere_curved_surface_area(radius):
    """Calculates the curved surface area of a hemisphere."""
    return 2 * math.pi * radius**2

def hemisphere_total_surface_area(radius):
    """Calculates the total surface area of a hemisphere."""
    return 3 * math.pi * radius**2

def hemisphere_volume(radius):
    """Calculates the volume of a hemisphere."""
    return (2/3) * math.pi * radius**3
'''
)
print("Module 'geometry_calculations.py' created successfully!")
```
