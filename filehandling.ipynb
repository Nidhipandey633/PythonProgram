```python
# Program for writing data into a file
try:
    # Opening file for write operation using 'with' statement for automatic closing
    with open("firstfile.txt", "w") as filev:

        # Writing five sentences into the file
        print("Enter any 5 sentences:")

        for x in range(5):
            # Input of data from user
            sentence = input(f"Sentence {x+1}: ")

            # Writing sentence into the file
            # We add + "\n" to ensure each sentence starts on a new line
            filev.write(sentence + "\n")
            print("--------------------")

    print("Data successfully written!")

except IOError as e:
    print(f"Error writing to file: {e}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
# to open file for read operation
try:
    with open("firstfile.txt","r") as filev:
        # to read data from file
        data = filev.read()
        print(data)
        print("-------------")
        print("No. of characters : ",len(data))
    # file is automatically closed here
except FileNotFoundError:
    print("Error: The file 'firstfile.txt' was not found.")
    print("Please ensure you have executed the cell that creates 'firstfile.txt' first.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
try:
    # Open the file in read mode
    with open("firstfile.txt", "r") as filev:
        content = filev.read()

    # Initialize vowel count
    vowel_count = 0

    # Define vowels (case-insensitive)
    vowels = "aeiouAEIOU"

    # Iterate through the content and count vowels
    for char in content:
        if char in vowels:
            vowel_count += 1

    print(f"The total number of vowels in 'firstfile.txt' is: {vowel_count}")

except FileNotFoundError:
    print("Error: 'firstfile.txt' not found. Please ensure you have run the cell that creates 'firstfile.txt' (cell oiY0kAbmsQTM) before running this cell.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
