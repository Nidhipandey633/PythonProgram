```python
import numpy as np

def get_even_factors(number):
    """Finds all even factors of a given number."""
    factors = []
    for i in range(1, int(number**0.5) + 1):
        if number % i == 0:
            if i % 2 == 0:
                factors.append(i)
            if (number // i) % 2 == 0 and (number // i) != i:
                factors.append(number // i)
    factors.sort()
    return np.array(factors)

# Example usage:
given_number = 60
even_factors_array = get_even_factors(given_number)
print(f"Even factors of {given_number}: {even_factors_array}")
```
